```python
# importing libraries

from bs4 import BeautifulSoup
import requests
import csv
```

```python
# connecting to the website

url = 'https://univ.mekari.com/collections?page=1'
page = requests.get(url).text
parsed = BeautifulSoup(page, 'html.parser')

# getting all the courses in the page

all_courses = parsed.find_all('li', {'class' : 'products__list-item'})
len(all_courses)
```

```python
titles = []
sale_prices = []
original_prices = []
categories = []
summaries = []
difficulties = []
materials = []
links = []

# connecting to the website

base_url = 'https://univ.mekari.com/collections?page='

for i in range(1,8) :
    url = base_url + str(i)
    print(f'Page {i}')
    print(url)
    page = requests.get(url).text
    parsed = BeautifulSoup(page, 'html.parser')

# getting all the courses in the page

    all_courses = parsed.find_all('li', {'class' : 'products__list-item'})

    for num, course in enumerate(all_courses) :

        # getting titles
        print(f'course {num+1}')
        title = course.find('span', {'class' : 'course-title'})
        titles.append(title.text)

        # getting discounted prices
        div_price = course.find('div', {'class' : 'course-card__details reset--details'})
        price = div_price.find_next('div')
        sale_prices.append(price.text.strip())

        # getting original prices
        ori_price = course.find('del')
        original_prices.append(ori_price.text.strip())

        # getting categories
        div_category = course.find('div', {'class' : 'category-wrap'})
        category = div_category.find_next('span', {'class' : 'tag'})
        categories.append(category.text)

        # getting summaries
        summary = course.find('p')
        summaries.append(summary.text)

        # getting difficulty levels
        difficulty = category.find_next('span', {'class' : 'tag'})
        difficulties.append(difficulty.text)
        
        # getting total materials
        material = course.find('span', {'class' : 'course-card__lesson-count'})
        materials.append(material.text)

        # getting links
        link = course.find('a', {'class' : 'course-card course-card__public published'})
        base_link = 'https://univ.mekari.com'
        links.append(base_link + link['href'])
```

```python
csv_header = ['course', 'discounted_price', 'original_price', 'category', 'summary', 'difficulty', 'total_materials', 'course_link']
with open('Mekari-University_Course_Web_Scraping.csv', 'w', encoding='UTF8', newline='') as file :
    writer = csv.writer(file)
    writer.writerow(csv_header)
    for i in range(len(titles)) :
        writer.writerow([titles[i], sale_prices[i], original_prices[i], categories[i], summaries[i], difficulties[i], materials[i], links[i]])
    
```

##### Let's take a look at the csv file

```python
import pandas as pd

df = pd.read_csv('Mekari-University_Course_Web_Scraping.csv')
df.head()
```

##### the discounted_price column is still in dollars ($) we need to convert it to rupiah

```python
df['discounted_price'] = df['discounted_price'].apply(lambda x: 'Rp0' if x == 'Gratis' else 'Rp' + x[1:] + '0')
```

##### same as the original price 

```python
# for the missing zeros
df['original_price'] = df['original_price'].apply(lambda x: 'Rp' + x[1:] + '0')
```

```python
# for the commas
df['original_price'] = df['original_price'].apply(lambda x: x.replace(',', '.') if ',' in x else x)
```
